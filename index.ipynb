## Question 1

what are the 4 main datatypes in SQLite3? Can we use other common types from other kinds of SQL?

 SQLite3 has four basic datatypes: TEXT, INTEGER, REAL, and BLOB (NULL not included). SQLite3 has a dynamic type system where values have their type associated with them and not their container like  other statically typed SQL database engines.  The dynamic type system is backwards compatible with the more common static type systems and allows storage of any value type in any column (except INTEGER PRIMARY KEY columns).  SQLite3 columns do not have rigid column typing but do have 'type affinities' which are recommendations for the type of values to be stored in the columns.  Type affinity allows common types from other kinds of SQL database engines to be used in SQLite3.  These common types are 'converted' into one of the four (five w/ NULL) basic datatypes allowing for compatibility between engines.

Primary keys are values in SQL tables which are unique identifiers for each record in the table.  Foreign keys are values in SQL tables that are references to primary keys in another table. 



## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

There are three different types of entity relationships in SQL databases: One-to-One, One-to-Many, and Many-to-Many.  In One-to-One relationships, each record in one table is related to one record in the other, and vice-versa.  In One-to-Many relationships, each record in the 'one' table can be related to many records in the 'many' table but each record in the 'many' table is related to only one record in the 'one' table.  In Many-to-Many relationships, each record in one table is related to many records in the other, and vice-versa.     


## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

There are four different types of 'joins' in SQL: inner, right, left, and full (outer).  Inner joins return all records that satisfy the join condition.  Right joins return all records from the right table and returns the records from the left table that satisfy the join condition.  Left joins are the opposite.  Full joins return all records from the right and left tables whether they satisfy the join conditions or not.
